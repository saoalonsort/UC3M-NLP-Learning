```python
import sklearn
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
text = ["El procesamiento de lenguaje natural es esencial hoy en día",
        "El procesamiento de lenguaje natural juega un papel muy importante hoy en día"]
```

```python
bigram_vectorizer = CountVectorizer(ngram_range=(2, 2))
```

```python
X = bigram_vectorizer.fit_transform(text)
print( bigram_vectorizer.get_feature_names_out())
```

```python
print(X.toarray())
```

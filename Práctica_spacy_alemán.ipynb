```python
import spacy
```

```python
!python -m spacy download de_core_news_sm
```

```python
nlp = spacy.load("de_core_news_sm")
```

```python
doc = nlp("Feierst du deinen Geburtstag in Deutschland? Oder kennst du einen Deutschen, der bald Geburtstag hat? Sicher freut er oder sie sich über ein Geburtstags-Lied auf deutsch. Heute zeige ich dir drei ganz verschiedene Lieder, die du an einem Geburtstag singen kannst.")
```

```python
document = nlp(doc)
```

```python
for i,s in enumerate(document.sents):
  print(i,s)
```

```python
for i, token in enumerate(document):
  print("1: ", token.orth_)
  print("2: ", token.lemma_)
  print("3: ", token.lower_)
  print("4: ", token.is_upper)
  print("5: ", token.suffix_)
  print("6: ", token.prefix_)
  print("7: ", token.pos_)
  print("8: ", token.shape_)
  print("9: ", token.is_alpha)
  print("10: ", token.dep_)
```

```python
for chunk in document.noun_chunks:
  print(chunk.text)
  print(chunk.root.text)
  print(chunk.root.dep_)
  print(chunk.root.head)
```

```python
for token in document:
  print(token.orth_)
  print(token.dep_)
  print(token.head)
  print(token.is_alpha)
  print(token.is_digit)
```

```python
from spacy import displacy
```

```python
displacy.render(document, style="dep", jupyter=True)
```

```python
displacy.render(document, jupyter= True, style="ent")
```

```python
for i,s in enumerate(document.sents):
  print(s)
  print("Entities: ")
  for ent in s.ents:
    print(ent.text, ent.label_, ent.start_char, ent.end_char)
```

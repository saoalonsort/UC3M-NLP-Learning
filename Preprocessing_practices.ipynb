```python
text = "He logrado resolver el problema gracias a la IA. Me ha sido de gran ayuda, ya que me ha ido diciendo todo lo que tenía que hacer. Tengo que reconocer que el mundo de la IA me resulta fascinante"
print(text)
```

```python
import spacy
nlp = spacy.load("es_core_news_sm")
```

```python
doc = nlp(text)
```

```python
lemmatized = [token.lemma_ for token in doc]
print("Lemmatized:", lemmatized)
```

```python
filtered = [token.text for token in doc if not token.is_stop and not token.is_punct]
print("Filtered:", filtered)
```

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
tokens = word_tokenize(text)
print("Tokens:", tokens)
```

```python
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word.lower() not in stop_words]
print("Without Stopwords:", filtered_tokens)
```
